```python
#Importing the libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
```

```python
data = pd.read_csv('Position_Salaries.csv')
print(data.head())
X = data.iloc[:, 1:2].values
y = data.iloc[:, 2].values
```

```python
reg_max_depth_tree = DecisionTreeRegressor(random_state=0, max_depth=3)
reg_max_depth_tree.fit(X, y)
```

```python
reg_min_depth_tree = DecisionTreeRegressor(random_state=0, min_samples_leaf=4)
reg_min_depth_tree.fit(X, y)
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
tree.plot_tree(reg_max_depth_tree, ax=ax, feature_names=['Level'], filled=True)
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
tree.plot_tree(reg_min_depth_tree, ax=ax, feature_names=['Level'], filled=True)
```

```python
from sklearn.tree import export_graphviz

export_graphviz(reg_max_depth_tree, out_file='tree.dot', feature_names=['Level'])
```

```python
plt.figure()
plt.scatter(X, y, marker='x', color='red', label='Data')
plt.plot(X, reg_max_depth_tree.predict(X), color='blue', label='max depth = 3')
plt.plot(X, reg_min_depth_tree.predict(X), color='green', label='max sample leaf = 4')
plt.title('Check it (Regression Model)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
